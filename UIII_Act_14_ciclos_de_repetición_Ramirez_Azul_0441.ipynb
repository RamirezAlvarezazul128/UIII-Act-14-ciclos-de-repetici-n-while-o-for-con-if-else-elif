## Problema 48 :
Programa que cuente vocales y consonantes en un texto, terminando la entrada al introducir un punto final .


```python
# --- CONTADOR DE VOCALES Y CONSONANTES REPETITIVO ---

while True:  # Ciclo para que se repita indefinidamente
    # Paso 1: Inicializamos los contadores en cero
    vocales = 0
    consonantes = 0

    # Paso 2: Entrada de datos interactiva
    texto = input("\nIngresa un texto (escribe 'salir' para cerrar el programa): ")

    # Condición para cerrar el programa completo
    if texto.lower() == "salir":
        print("¡Programa terminado!")
        break

    # Paso 3: Ciclo para recorrer cada letra del texto
    for letra in texto:

        # Condición para terminar la frase al introducir un punto final
        if letra == ".":
            break

        # Convertimos la letra a minúscula
        letra = letra.lower()

        # Comprobamos si es una vocal
        if letra == "a" or letra == "e" or letra == "i" or letra == "o" or letra == "u":
            vocales = vocales + 1

        # Comprobamos si es una consonante
        elif letra >= "a" and letra <= "z":
            consonantes = consonantes + 1

    # Paso 4: Mostramos los resultados finales en pantalla
    print("--- RESULTADOS ---")
    print("Total de vocales encontradas:", vocales)
    print("Total de consonantes encontradas:", consonantes)
    print("programa realizado por Ramirez Azul NL = 48")
```
